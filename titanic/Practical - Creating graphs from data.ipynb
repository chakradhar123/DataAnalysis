```python
import matplotlib.pyplot as plt
#Effect of time spent walking (hours) on the distance travelled (miles)
time_spent_walking = [1,2,3,4,5]
distance = [2,4,6,8,10]
plt.plot(time_spent_walking,distance)
plt.show()
```
