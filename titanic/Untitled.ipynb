```python
def getTitle(name):
    if '.' in name:
        return name.split(',')[1].split('.')[0].strip()
    else:
        return 'no title'
```

```python
titles=set([x for x in df.Name.map(lambda x:getTitle(x))])
```

```python
titles
```

```python
def smaller_titles(x):
    title=x['title']
    if title in ['Capt','Col','Major']:
        return 'Officer'
    elif title in ['Jonkheer','Lady','Don','Sir','the Countess']:
        return 'Royal'
    elif title=='Mme':
        return 'Mrs'
    elif title in ['Mlle','Ms']:
        return 'Miss'
    else:
        return title
    
        
```

```python
df['title']=df.Name.map(lambda x:getTitle(x))
```

```python
df
```

```python
df['title']=df.apply(smaller_titles,axis=1)
```

```python
df
```

```python
df.drop('Name',axis=1,inplace=True)
```

```python
df.drop('Ticket',axis=1,inplace=True)
df.sample(20)
```

```python
df.Sex.replace(['male','female'],[0,1],inplace=True)
df.Embarked.replace(['S','C','Q'],[0,1,2],inplace=True)
df.title.replace(['Mr','Miss','Mrs','Master','Dr','Rev','Royal','Officer'],[0,1,2,3,4,5,6,7],inplace=True)
```
